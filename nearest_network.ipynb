```python
clinton_corpus = df_clinton[:][0].str.cat(sep='\n')
trump_corpus = df_trump[:]['text'].str.cat(sep='\n')
```

```python
clinton_sentences = clinton_corpus.split('\n')
trump_sentences = trump_corpus.split('\n')
clinton_sentences = [line.lower().split(' ') for line in clinton_sentences]
trump_sentences = [line.lower().split(' ') for line in trump_sentences]
```

```python
def clean(s):
    return [w.strip(',.#"!?:;()\'') for w in s if len(w)>2 and 'http' not in w and '@' not in w and '#' not in w]
clinton_sentences = [clean(s) for s in clinton_sentences if len(s) > 0]
trump_sentences = [clean(s) for s in trump_sentences if len(s) > 0]
```

```python
from gensim.models import Word2Vec

clinton_model = Word2Vec(clinton_sentences, size=100, window=5, min_count=3, workers=4)
trump_model = Word2Vec(trump_sentences, size=100, window=5, min_count=3, workers=4)
```

```python
clinton_vectors = clinton_model.wv
trump_vectors = trump_model.wv
```

```python
ordered_clinton = [(term, voc.index, voc.count) for term, voc in clinton_model.wv.vocab.items()]
ordered_clinton = sorted(ordered_clinton, key=lambda k: -k[2])
ordered_terms_c, term_indices_c, term_counts_c = zip(*ordered_clinton)
word_vectors_c = pd.DataFrame(clinton_model.wv.syn0[term_indices_c, :], index=ordered_terms_c)
word_vectors_c[:10]
```

```python
ordered_trump = [(term, voc.index, voc.count) for term, voc in trump_model.wv.vocab.items()]
ordered_trump = sorted(ordered_trump, key=lambda k: -k[2])
ordered_terms_t, term_indices_t, term_counts_t = zip(*ordered_trump)
word_vectors_t = pd.DataFrame(trump_model.wv.syn0[term_indices_t, :], index=ordered_terms_t)
word_vectors_t[:10]
```

```python
def related_terms(model, token, topn=10):
    out = []
    for word, similarity in model.most_similar(positive=[token], topn=topn):
        out.append([word, round(similarity, 3)])
    return out
print(related_terms(clinton_model, 'hillary'), '\n')
print(related_terms(trump_model, 'trump'))
```

```python
def network(model, token, topn=10):
    base = related_terms(model, token, topn)
    network = []
    for i in range(topn, 0, -1):
        new_related = related_terms(model, base[topn - i][0], i)
        network.append([new_related, [0], [topn - i]])
        
        for layer in range(i//2, 0, -1):
            layered_related = related_terms(model, new_related[i//2 - layer][0], layer)
            network.append([layered_related, [i//2 - layer + 1], [topn - i]])
    return network

clinton_network = network(clinton_model, 'clinton')
for words in clinton_network:
    for items in words: 
        for word in items:
            print(word)
    print('\n')
```

```python
trump_network = network(trump_model, 'trump')
for words in trump_network:
    for items in words: 
        for word in items:
            print(word)
    print('\n')
```
